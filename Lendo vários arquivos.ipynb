```python
import pandas as pd
import numpy as np
import json
import os
```

```python
directory = 'dataset'
df = pd.DataFrame(columns=["Id",'StartTime','EndTime','StartLAT','StartLONG',"Distance"])
for filename in os.listdir(directory):
    if filename.endswith(".json"):
        file = open(str(directory+"/"+filename))
        data = json.load(file)
        dici = {}
        dici["Id"] = filename[27:47]
        print(dici["Id"])
        if "distance" in data:
            dici['Distance'] = data["distance"]
        else:
            dici["Distance"] = np.nan
        if "latitude" in data:
            dici["StartLAT"] = data["latitude"]
        else:
            dici["StartLAT"] = np.nan
        if "longitude" in data:
            dici["StartLONG"] = data["longitude"]
        else:
            dici["StartLONG"] = np.nan
        if "startTime" in data:
            dici["StartTime"] = data["startTime"]
        else:
            dici["StartTime"] = np.nan
        if "stopTime" in data:
            dici["EndTime"] = data["stopTime"]
        else:
            dici["EndTime"] = data["stopTime"]
        df = df.append(dici, ignore_index=True)
```

```python
df['StartTime'] = pd.to_datetime(df["StartTime"])
df['EndTime'] = pd.to_datetime(df["EndTime"])
```
